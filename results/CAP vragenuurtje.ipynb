```python
def top10(row, n=10):
    #print list(row.sort_values(ascending=False).index[:10])
    return pd.Series(list(row.sort_values(ascending=False).index[:n]))

config = {
    "inputData": "/path/to/input/data/*",
    "outDir": "/home/jvdzwaan/data/dilipad/CAP/{}",
    "nTopics": 100
}

vragenuurtje_topics = pd.read_csv(config.get('outDir').format('theta_100.csv'), index_col=0, encoding='utf-8')
# add top 10 of topics estimated in text
vragenuurtje_topics[['top1', 'top2', 'top3', 'top4', 'top5', 'top6', 'top7', 'top8', 'top9', 'top10']] = vragenuurtje_topics.apply(lambda row: top10(row), axis=1)
#vragenuurtje_topics
```

```python
config = {
    "inputData": "/path/to/input/data/*",
    "outDir": "/home/jvdzwaan/data/dilipad/results/all_years-adj-cabinets_selected-parties/{}",
    "nTopics": 100
}

usefulness_data = pd.read_excel(config.get('outDir').format('topics_100.xlsx'), index_col=0)
```

```python
cap_results = pd.DataFrame(index=vragenuurtje_topics.index, columns=['question_id', 'manual', 'topicmodeling1'])
for text, topic in vragenuurtje_topics['top1'].iteritems():
    #print 'text', text
    question_id = int(text.replace('.txt', '')) -1
    #print 'topic no', topic
    #print 'cap category', usefulness_data.loc[int(topic), 'cap']
    #print 'coded cap category', data.loc[question_id, 'main code']
    #print
    cap_results.loc[text, 'question_id'] = question_id
    cap_results.loc[text, 'manual'] = data.loc[question_id, 'main code']
    cap_results.loc[text, 'topicmodeling1'] = usefulness_data.loc[int(topic), 'cap']
print np.sum(cap_results['manual']==cap_results['topicmodeling1'])
print np.sum(cap_results['manual']==0)
print np.sum(cap_results['topicmodeling1']==0)
print np.sum(cap_results['topicmodeling1'].isnull())
#cap_results
```

```python
def get_topic_cap_results(series, usefulness_data):
    result = pd.Series(index=series.index)
    for text, topic in series.iteritems():
        result[text] = usefulness_data.loc[int(topic), 'cap']
    return result

cap_results['topicmodeling2'] = get_topic_cap_results(vragenuurtje_topics['top2'], usefulness_data)
cap_results['topicmodeling3'] = get_topic_cap_results(vragenuurtje_topics['top3'], usefulness_data)
cap_results['intop3'] = cap_results.apply(lambda row: row['manual'] if row['manual'] in [row['topicmodeling1'], row['topicmodeling2'], row['topicmodeling3']] else 0, axis=1)
cap_results['intop3binary'] = cap_results.apply(lambda row: row['manual'] in [row['topicmodeling1'], row['topicmodeling2'], row['topicmodeling3']], axis=1)
cap_results
```

```python
import glob

d = pd.read_csv('/home/jvdzwaan/data/dilipad/CAP/thetas/theta_100_0.csv', index_col=0)
res = pd.DataFrame(index=d.index)

files = glob.glob('/home/jvdzwaan/data/dilipad/CAP/thetas/*_*_*.csv')
print files
for i, fi in enumerate(files):
    df = pd.read_csv(fi, index_col=0)
    res[i] = df.apply(lambda row: top10(row, 1), axis=1)
print topic2cap
res[['cap{}'.format(i) for i in range(10)]] = res.apply(lambda series: [topic2cap[int(v)] for v in series])
print res
```

```python
from collections import Counter

def vote(row):
    c = Counter()
    for i in row[['cap{}'.format(i) for i in range(10)]]:
        c[i] += 1
    return pd.Series(list(c.most_common()[0]))

res = res.fillna(0)
res[['most_frequent_cap', 'freq']] = res.apply(lambda row: vote(row), axis=1)
res
```
