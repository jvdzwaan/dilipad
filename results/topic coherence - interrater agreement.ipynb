In \cite{Aletras2013}: 
Inter-annotator agreement (IAA) is measured as the average of the Spearman correlation between the
set of scores of each survey respondent and the average of the other respondents’ score.

```python
import pandas as pd
# load data

data = pd.read_excel('/home/jvdzwaan/Desktop/topics_100.xlsx')
data
```

```python
import numpy as np
from scipy.stats import pearsonr

print np.mean(data[['usefulness_B', 'usefulness_C']].values, axis=1)

print pearsonr(data['usefulness_A'].values, np.mean(data[['usefulness_B', 'usefulness_C']].values, axis=1))
```

```python
columns = ['usefulness_A', 'usefulness_B', 'usefulness_C']

results = []
for c in columns:
    others = columns[:]
    others.remove(c)
    print c, others
    p, sign = pearsonr(data[c].values, np.mean(data[others].values, axis=1))
    if sign >= 0.05:
        print 'not significant', c
    results.append(p)
print results
print np.mean(results)
```
